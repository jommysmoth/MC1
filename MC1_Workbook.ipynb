```python
data = pd.read_csv('Lekagul Sensor Data.csv')
data_car_id = data['car-id'].as_matrix()
data.columns
plot_width = 500
plot_height = 500
```

```python
data['car-int-id'] = pd.factorize(data['car-id'])[0]+1
# Replacing unique car id with factorized integer for ease
data['gate-name-int'] = pd.factorize(data['gate-name'])[0] + 1
ts = data['Timestamp'].values
string_date = []
```

```python
for iter, timedata in enumerate(ts):
    time_tuple = datetime.strptime(timedata, "%Y-%m-%d %H:%M:%S")
    time_string = time_tuple.strftime('%Y %m %d')
    string_date.append(time_string)    
```

```python
string_dates = np.asarray(string_date)
```
